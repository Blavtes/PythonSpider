```python
# 行操作  loc 显示索引  / iloc 隐式索引
df= pd.DataFrame({'Country':['China','China','India','India',"America",'America'],
                 'Income':[10000,1000,5000,5002,4000,2003],
                 'Age':[50,53,45,12,34,56]})
```

```python
df.loc[[0]]
```

```python
df.iloc[[0,3]]
```

```python
df.loc[:,'Country'] # 所以行，指定列 》》》 等同于df['Country']
```

```python
df.loc[[2,4],['Country','Age']]
```

```python
df.iloc[[2,4],[0,2]] #隐式取值
```

```python
# 切片
df.iloc[0:3,0:3]
```

```python
df[df.Country=='India'].iloc[0:2]
```

```python
# age 列等于偶数的那些数据，赋值90
df[df['Age']%2==0]['Age']=90
```

```python
df[df['Age']%2==0].index
```

```python
df.loc[df[df['Age']%2==0].index,'Age']=90
```
