```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},
                 index=[0,1,2,3])
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                  'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7'],
                  'D':['D4','D5','D6','D7']},
                 index=[4,5,6,7])
                 
df3=pd.DataFrame({'B':['B2','B3','B6','B7'],
                  'D':['D2','D3','D6','D7'],
                  'F':['F2','F3','F6','F7']},
                 index=[2,3,6,7])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],keys=['x','y'])
```

```python
pd.concat([df1,df3],axis=1) # 默认取并集
```

```python
pd.concat([df1,df3],axis=1,join='inner') # 取交集
```

```python
# ignore_index 忽略索引、重置index
pd.concat([df1,df3],ignore_index=True)
```

```python
pd.concat([df1,df3]).reset_index()
```

```python
pd.concat([df1,df3]).reset_index(drop=True)
```

```python
#drop 和concat axis=0，行，axis=1列
#其他的函数相反


#merge函数

#merge和concat的区别，
#1、concat可以实现多张表的拼接，merge只能完成2张表的连接
#2、pd.merge() 会自动根据column名称相同的列进行拼接
#3、concat拼接的时候不需要key列
df1 = pd.DataFrame({'name':['Bobs','Linda','Bill'],
                    'group':['Accounting','Product','Marketing'],
                    'hire_date':[1998,2017,2018]})
df2=pd.DataFrame({'name':['James','Bobs','Bill'],'hire_dates':[1998,2016,2007]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='outer') #nan值补全
```

```python
pd.merge(df1,df2,how='left')# 以左表为基础
```

```python
# 一对一合并
df1=pd.DataFrame({'employee':['Bob','Jake','Lisa'],
                 'group':['Accounting','Enginnering','Enginnering']})
df2=pd.DataFrame({'employee':['Lisa','Bob','Jake'],
                 'hire_date':[2004,2008,2012]})
df3=pd.DataFrame({'employee':['Lisa','Bob','Jake'],
                 'hire_date':[2004,2003,2002]})

df5=pd.DataFrame({'employee':['Bob','Jake','Lisa'],
                 'group':['Accounting','Enginnering','Enginnering'],
                 'hire_date':[2002,2023,2012]})
df4=pd.DataFrame({'employee':['Lisa','Bob','Jake'],
                 'hire_date':[2012,2002,2002]})
```

```python
df1
```

```python
df2
```

```python
# on 参数
pd.merge(df1,df2,on='employee')
```

```python
pd.merge(df2,df3,on='hire_date')
```

```python
# 多对多合并
pd.merge(df2,df3,how='right')
```

```python
pd.merge(df3,df2,how='right')
```

```python
pd.merge(df4,df5,on='hire_date')
```

```python
pd.merge(df4,df5,on='hire_date',suffixes=['_left','_right'])
```

```python
# right_index参数
df4
```

```python
df5
```

```python
pd.merge(df4,df5,left_on='employee',right_index=True)
```

```python
# map是Series的方法，只能被 Series调用，一般和函数联用,map不能传递多个参数
#apply 可以应用于行和列，也可以应用于dataframe，可以传递过个参数
# 它们 通常和函数联用

df1=pd.DataFrame({'employee':['Bob','Jake','Lisa'],
                 'group':['Accounting','Enginnering','Enginnering'],
                 'gender':['男','女','男'],
                 'age':[43,80,21]})
df1
```

```python
def apply_age(x,bias):
    return x+bias

df1['age'].apply(apply_age,bias=2)
```

```python

df1=pd.DataFrame({'employee':['Bob','Jake','Lisa'],
                 'group':['Accounting','Enginnering','Enginnering'],
                 'gender':['男','女','男'],
                 'weight':[61,45,55],
                  'height':[172,161,168],
                 'age':[43,80,21]})
df1
```

```python
df1[['height','weight','age']].apply(sum,axis=0)
```

```python
df1['height'].sum()
```

```python
def wh(x):
    weight=x['weight']
    height=x['height']
    wh = weight/height
    return wh

df1['wh']=df1.apply(wh,axis=1)
```
