```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = 'Arial Unicode MS' # mac
plt.rcParams['axes.unicode_minus'] = False # 特殊字符也显示
```

```python
value = list(range(10))
category = list('qwertyuiop')
```

```python
category
```

```python
fig = plt.figure(figsize=(8,4))
# 垂直柱状图
ax1 = fig.add_subplot(121)
ax1.set_title('图1')
ax1.bar(x = category,height = value,
       alpha=0.5,
       width=0.5,
       color='yellow',
       edgecolor='r',
       linewidth=2)
#水平柱状图
ax2 = fig.add_subplot(122)
ax2.set_title('水平')
ax2.barh(y=category,width=value,
       alpha=1,
       height=0.5,
       color=['yellow','red','blue','grey','magenta'],
       )

plt.show()
```

```python
# 堆叠柱状图
x = ['周一','周二','周三','周四','周五','周六','周日']
y1 = [7,6,5,4,3,2,1]
y2 = [1,2,3,2,1,1,0.5]

y3 = [0.1,0.2,0.9,0.4,0.5,0.5,0.25]

plt.figure(figsize=(10,7))
# 一个图表上多个数据系列
plt.bar(x,y1,label='系列1',edgecolor='black')
plt.bar(x,y2,label='系列2',edgecolor='blue')
# plt.bar(x,y3,label='系列3',edgecolor='blue')

#设置图例
plt.legend(loc=1,fontsize=12)
plt.title('堆叠柱状图',fontsize=16)
plt.xlabel('星期',fontsize=16)
plt.ylabel('我是y轴',fontsize=16)

# 标签
# 文本标记 zip> plt.text(x,y,文本值)
for x1,y1,y2 in zip(x,y1,y2):
    p1 = y1/(y1+y2)
    p2=y2/(y1+y2)
    plt.text(x1,y2+(y1-y2)*0.4,'{:.0%}'.format(p1),ha='center',fontsize=15)
    plt.text(x1,y2*0.4,'{:.0%}'.format(p2),ha='center',fontsize=15)

    
plt.show()
```

```python
# 分组柱状图
x= list(range(7))
xticks1 = ['周一','周二','周三','周四','周五','周六','周日']
y1 = [7,6,5,4,3,2,1]
y2 = [1,2,3,3,1,1,0.5]

plt.figure(figsize=(9,6))
plt.bar(x,y1,width=0.4,label='系列1',edgecolor='r',alpha=0.7)
plt.bar([x[i]+0.4 for i in x],y2,width=0.4,label='系列2',edgecolor='r',alpha=0.7)


plt.xticks(x,xticks1)
plt.legend(loc=1,fontsize=12)
plt.title('分组柱状图',fontsize=16)
plt.xlabel('星期',fontsize=16)
plt.ylabel('我是y轴',fontsize=16)

plt.show()
```

```python
# array 可直接计算
x = np.array(list(range(7)))

xticks1 = ['周一','周二','周三','周四','周五','周六','周日']
y1 = [7,6,5,4,3,2,1]
y2 = [1,2,3,3,1,1,0.5]

plt.figure(figsize=(9,6))
plt.bar(x,y1,width=0.4,label='系列1',edgecolor='r',alpha=0.7)
plt.bar(x+0.4,y2,width=0.4,label='系列2',edgecolor='r',alpha=0.7)


plt.xticks(x,xticks1)
plt.legend(loc=1,fontsize=12)
plt.title('分组柱状图',fontsize=16)
plt.xlabel('星期',fontsize=16)
plt.ylabel('我是y轴',fontsize=16)
plt.show()
```

```python
# 直方图 频数、频率分布

#组数、组距 （max - min）/组距 = 组数

data = np.random.randn(1000)
# hist
plt.hist(data)

plt.show()
```

```python
%matplotlib inline
plt.style.use('seaborn-white')

data = np.random.randn(1000)
# hist
plt.hist(data,alpha=0.5,density=True,color='g')
plt.hist(data,bins=30,density=True,alpha=0.5,color='g')

plt.show()
```

```python
plt.rcParams['font.sans-serif'] = 'Arial Unicode MS' # mac

time= [161,98,151,123,126,165,134,156,143,128,111,132,128,107,106]
plt.figure(figsize=(10,5),dpi=100)
#组距
distance = 2
group_num = int((max(time)-min(time))/distance)
plt.hist(time,bins=group_num)
plt.xticks(range(min(time),max(time))[::2],fontsize=8)
plt.grid(linestyle='--',alpha=0.5)
plt.xlabel("电影时长")
plt.ylabel("电影数据量")
plt.show()
```
