```python

data = {'dataset':['I','I','I','II','II','II','III','III','III','IV','IV','IV'],
       'x':np.random.rand(12) * 10,
       'y':np.random.rand(12)}
data
df = pd.DataFrame(data)
df
```

```python
#第一种接口
# 图1 上的数据
data1= df[df['dataset']=='I']
data2= df[df['dataset']=='II']
data3= df[df['dataset']=='III']
data4= df[df['dataset']=='IV']

plt.figure(figsize=(10,8))

plt.subplot(221)
plt.plot(data1['x'],data1['y'],'o') # 数据点标记
plt.title('data1')

plt.subplot(222)
plt.plot(data2['x'],data2['y'],'o') # 数据点标记
plt.title('data2')

plt.subplot(223)
plt.plot(data3['x'],data3['y'],'o') # 数据点标记
plt.title('data3')

plt.subplot(224)
plt.plot(data4['x'],data4['y'],'o') # 数据点标记
plt.title('data4')

plt.show()
```

```python
# 第二种接口 面向对象
fig= plt.figure(figsize=(10,8))

ax1=fig.add_subplot(221) #构建2行2列画图区， 221 在第一个子图区绘图

ax2=fig.add_subplot(222) 
ax3=fig.add_subplot(223) 
ax4=fig.add_subplot(224) 

ax1.plot(data1['x'],data1['y'],'o') # 数据点标记
ax2.plot(data2['x'],data2['y'],'o') # 数据点标记
ax3.plot(data3['x'],data3['y'],'o') # 数据点标记
ax4.plot(data4['x'],data4['y'],'o') 

ax1.set_title('data1')
ax2.set_title('data2')
ax3.set_title('data3')
ax4.set_title('data4')

fig.suptitle('Ancombe data')
fig.tight_layout()# 紧凑布局

plt.show()
```

```python
# 折线图
fig = plt.figure()
ax = fig.add_subplot()
x = ['S','M','L']
y = np.array([1,3,5])
ax.plot(x,y)

ax.plot(x,y+3)
plt.show()
```

```python
# linestyle

fig= plt.figure()
ax = fig.add_subplot()
x = ['S','M','L']
y = np.array([1,3,5])
ax.plot(x,y)
ax.plot(x,y+3,color='b',linestyle='-')
ax.plot(x,y+2,color='r',linestyle='--')

ax.plot(x,y+1,color='#FFDD44',linestyle='-.')
ax.plot(x,y+0.5,linestyle=':')

plt.show()
```

```python
# 坐标轴范围
fig= plt.figure()
ax = fig.add_subplot()
x = np.array([0,1,3])
y = np.array([1,3,5])
ax.plot(x,y+3,'-.k')
ax.plot(x,y+2,'-g')

# 范围
plt.xlim(0,3)
plt.ylim(3,10)
plt.show()
```

```python
# 特殊标记点

fig= plt.figure()
ax = fig.add_subplot()
x = np.array([0,1,3])
y = np.array([1,3,5])
ax.plot(x,y+3,'-.k')
ax.plot(x,y+2,'-g',marker='v') # 查参数

plt.show()
```

```python
import matplotlib.pylab as plt
plt.rcParams['font.sans-serif'] = 'Arial Unicode MS' # mac
plt.rcParams['axes.unicode_minus'] = False # 特殊字符也显示
```

```python
plt.figure()
x = list(range(1,8))
y_zz = [15,20,22,23,20,18,16]
y_cc = [-10,-8,-12,-10,-8,-6,-1]

plt.plot(x,y_zz,linestyle=':',color='r',marker='d',markersize=6,markerfacecolor='b')

plt.plot(x,y_cc,linestyle='-.',color='b',marker='d',markersize=6,markerfacecolor='r')
plt.xlabel(' 日期')

plt.ylabel('温度(°C)')
plt.title(' 一周温度变化趋势')
plt.legend(['郑州温度','长春温度'],loc='best')

# 修改xy - tick的值
xticks = ['周一','周二','周三','周四','周五','周六','周日']

plt.xticks(x,xticks,rotation=90)
#数据点标签 zip(x,y)
for i,j in zip(x,y_zz):
    plt.text(i,j + 1,'%d°C' % j,ha ='center')

for i,j in zip(x,y_cc):
    plt.text(i,j + 1,'%d°C' % j,ha ='center')

plt.show()
```
