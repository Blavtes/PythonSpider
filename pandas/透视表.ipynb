```python
# avg_salary 不同公司的平均薪资

import numpy as np
company=['A','B','C']
data = pd.DataFrame({
    'company':[company[x] for x in np.random.randint(0,len(company),10)],
    "salary":np.random.randint(5,50,10),
    'age':np.random.randint(15,50,10)
})
data
```

```python
data.groupby('company').get_group('A')
```

```python
data.groupby('company')['salary'].mean()
```

```python
dic =data.groupby('company')['salary'].mean().to_dict()
```

```python
data['avg_salary'] = data['company'].map(dic)
```

```python
data
```

```python
data['avg_salary']=data.groupby('company')['salary'].transform('mean')
```

```python
data
```

```python
# 找平均薪水大于25的
data.groupby('company')['salary'].mean()
```

```python
def func(x):
    return x['salary'].mean()>25

data.groupby('company').filter(func)
```

```python
#lambda 匿名函数
data.groupby('company').filter(lambda x:x['salary'].mean()>25)
```

```python
# apply
data.groupby('company').apply(func)
```
