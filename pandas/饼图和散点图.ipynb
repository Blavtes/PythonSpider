```python
import matplotlib.pyplot as plt
plt.pie(x=[1,2,3],labels=['A','B','C'],colors=plt.cm.get_cmap('Set3_r')(range(3)),
        explode=(0,0.1,0),
        autopct='%.2f%%',
        pctdistance=0.4,
        labeldistance=0.5,
        startangle=45,
        wedgeprops={
            'edgecolor':'r',
            'linestyle':'-.',
            'alpha':0.5},
        textprops={
           'fontsize':16,
            'color':'r',
            'alpha':0.5} 
       )
#  Set3_r -> Set3 颜色反转
# labeldistance 标签离圆心的位置
# pctdistance 数值偏离圆心位置
# startangle 旋转
plt.axis('equal')
plt.show()
```

```python
import random
x1 = [random.randint(1,1000) for i in range(100)]
y1 = [random.randint(1,1000) for i in range(100)]
x2 = [random.randint(1,1000) for i in range(100)]
y2 = [random.randint(1,1000) for i in range(100)]
```

```python
colors1 = '#00ced1'
colors2 = '#DC143C'
area  = 3.14 * 4 ** 2

plt.scatter(x1,y1,s = area,c=colors1,alpha=0.4,label='类别A')
plt.scatter(x2,y2,s = area,c=colors2,alpha=0.4,label='类别B')

plt.legend(loc=1)

plt.show()
```
